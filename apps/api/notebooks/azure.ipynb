```python
from typing import Any, Dict, Literal, Optional
from pydantic import BaseModel
import requests
def test_azure_path(base_url, path, times = 100):
    data = {}
    url = base_url + path
    for i in range(times):
        response = requests.get(url)
        if response.status_code != 200:
            print("Error: ", response.status_code)
        else:
            res_json = response.json()
            print("Request: ",i, res_json)
            for k in res_json:
                if k not in data:
                    data[k] = []
                data[k].extend(res_json[k])
    return data

class TestUrl(BaseModel):
    url: str
    method: Literal["GET", "POST"] 
    request_body: Optional[Dict[str, Any]] = None

def test_azure_url(base_url, test_url_body:TestUrl,  times = 100):
    data = {}
    url = base_url + "/test/test-url"
    for i in range(times):
        response = requests.post(url, json=test_url_body.dict())
        if response.status_code != 200:
            print("Error: ", response.status_code)
        else:
            res_json = response.json()
            print("Request: ",i, res_json)
            for k in res_json:
                if k not in data:
                    data[k] = []
                data[k].append(res_json[k])
    return data
```

```python

import numpy as np
def stats(data):
    """Data should be a dictonary mapping key to list of numbers"""
    for k in data:
        if len(data[k]) == 0:
            continue
        print(k.upper(), np.mean(data[k]), np.std(data[k]))
        print(np.min(data[k]),np.percentile(data[k],25) , np.median(data[k]), np.percentile(data[k],75),np.percentile(data[k],90), np.percentile(data[k],95), np.max(data[k]))
        print()
```

```python
import asyncio
import requests

def fetch(url, i):
    response = requests.get(url)
    if response.status_code != 200:
        print("Error: ", response.status_code)
        return None
    res_json = response.json()
    print("Request:", i, res_json)
    return res_json

async def test_azure_async(base_url, n=10, times=100):
    data = {}
    path = "/test/azure-openai-embed?embed_times=0&embed_inv_times=5&cache=no-cache&print_error=true"
    url = base_url + path

    async def fetch_async(i):
        return await asyncio.to_thread(fetch, url, i)

    tasks = [fetch_async(i) for i in range(times)]
    responses = await asyncio.gather(*tasks)

    for res_json in responses:
        if res_json:
            for k in res_json:
                if k not in data:
                    data[k] = []
                data[k].extend(res_json[k])

    return data

```

```python
import socket
import ssl
import time
from urllib.parse import urlparse
import json

def measure_full_request_time(url, method='GET', body=None):
    # Parse the URL to extract hostname, port, and path
    parsed_url = urlparse(url)
    hostname = parsed_url.hostname
    port = parsed_url.port if parsed_url.port else (443 if parsed_url.scheme == 'https' else 80)
    path = parsed_url.path if parsed_url.path else '/'
    
    # Construct the HTTP request
    request_line = f"{method} {path} HTTP/1.1\r\nHost: {hostname}\r\n"
    headers = "Connection: close\r\n\r\n"
    if isinstance(body, dict):
        body = json.dumps(body)
    request_body = body if body else ''
    request = request_line + headers + request_body
    
    full_start = time.time()

    # Measure DNS resolution time
    dns_start = time.time()
    ip_address = socket.gethostbyname(hostname)
    dns_end = time.time()

    # Measure TCP handshake time
    tcp_start = time.time()
    sock = socket.create_connection((ip_address, port))
    tcp_end = time.time()

    # Measure TLS handshake time (only if HTTPS)
    if parsed_url.scheme == 'https':
        tls_start = time.time()
        context = ssl.create_default_context()
        sock = context.wrap_socket(sock, server_hostname=hostname)
        tls_end = time.time()
    else:
        tls_start = tls_end = tcp_end  # No TLS handshake for HTTP

    # Measure total time for full request (HTTP request and data transfer)
    total_start = time.time()
    sock.sendall(request.encode())
    response = sock.recv(4096)
    total_end = time.time()
    
    # Close the connection
    sock.close()
    
    data = {
        "dns": ((dns_end - dns_start) * 1000),
        "tcp": ((tcp_end - tcp_start) * 1000),
        "tls": ((tls_end - tls_start) * 1000) if parsed_url.scheme == 'https' else 0,
        "http": ((total_end - total_start) * 1000),
        "full": ((total_end - full_start) * 1000),
    }

    print("DNS time: ", data["dns"], "ms")
    print("TCP time: ", data["tcp"], "ms")
    print("TLS time: ", data["tls"], "ms")
    print("HTTP time: ", data["http"], "ms")
    print("Full time: ", data["full"], "ms")

    return data

measure_full_request_time('https://develop-api.syntag.ai/vapi/server-url', method='POST', body={'url': 'https://develop-api.syntag.ai/test/database'})
```
